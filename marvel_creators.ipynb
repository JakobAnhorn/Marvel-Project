```python
#using api reuqest and for loops to pull movie authors, sales and writers
writer=[]
sales=[]
title=[]
list = sorted_df["movie_title"]

for i in list:
    data = requests.get(url + i + api_key).json()
    author = data["Writer"]
    box_office = data["BoxOffice"]
    tag = data["Title"]
    writer.append(author)
    sales.append(box_office)
    title.append(tag)
print(writer)
print(sales)
print(title)
```
